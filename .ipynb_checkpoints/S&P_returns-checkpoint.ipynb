```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8")
import yfinance as yf
import datetime

from pandas import DataFrame, Series
from pandas_datareader import data as pdr

# Needs to be done in order for pandas yahoo finance to work
yf.pdr_override()
```

```python
stocks = ['SPY']
every_month_inv = 100
years = 10
```

```python
def get_data(stocks, start, end):
    return pdr.get_data_yahoo(stocks, start, end) 
```

# S&P Investment

In the world of finance, the S&P 500 stands tall as a beacon of opportunity for investors seeking long-term growth and financial prosperity. This renowned stock market index, comprised of 500 of the largest publicly traded companies in the United States, has consistently delivered an average annual return of 6-7% over the years. Investing in the S&P 500 presents a compelling alternative to traditional savings methods, offering the potential for substantial returns and financial security.

Today, we will explore the hypothetical scenario of an investment strategy involving the S&P 500 over the past 10 years. Specifically, we will consider the potential outcomes of committing $100 per month to this investment. It is essential to note that the designated investment timeframe, as well as the monthly allocation, can be changed.

```python
start = (datetime.datetime.now() - datetime.timedelta(days=365*years)).replace(hour=0, minute=0, 
                                                                            second=0, microsecond=0)
end = datetime.datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)

data = get_data('SPY', start, end)
plt.plot(data['Adj Close'])
plt.title('S&P 500')
plt.xlabel('Time')
plt.ylabel('Price')
plt.show()
```

# Investment Performance Analysis

Upon a decade-long commitment to the S&P 500 with a monthly contribution of $100, the resultant earnings exhibit an impressive nearly 100 percent growth on the initial investment. It is imperative to note that dividends have not been factored into this analysis. This observation underscores the favorable returns associated with investing in the S&P, substantiating its superiority over conventional savings methods.

```python
data['time_start'] = data.index
data['time_start'] = data['time_start'].apply(lambda x : pd.to_datetime(str(x).split(' ')[0]))
data.index = data['time_start']

date_range = pd.date_range(start, end, freq='BMS')
date_range = date_range[date_range.isin(data.index)]

inv_periods = data.loc[date_range]['Adj Close']
current_price = data.iloc[-1]['Adj Close']

invest_points = current_price / inv_periods
invest_points = every_month_inv * invest_points

investment = round(invest_points.sum(), 2)
money_inv = (every_month_inv * len(date_range))
perc_inv = investment / money_inv
perc_inv_ann = perc_inv**(1/years)

perc_inv = round((perc_inv - 1)*100, 1)
perc_inv_ann = round((perc_inv_ann - 1)*100, 2)

print("Investment:", investment, "Total money invested:", money_inv, "\nPercentage return", perc_inv, 
     "anually:", perc_inv_ann)
```

# Analysis of Daily Market Fluctuations

The day-to-day fluctuations in the market are inherently characterized by their random and noisy nature. Consequently, for individuals without a professional background in trading, navigating the intricacies of the market proves to be a challenging path to becoming a successful market participant and achieving consistent profitability in the long run.

For the majority of investors, adopting a long-term investment approach emerges as the most prudent strategy. The inherent risk associated with a well-considered long-term investment is notably minimal, offering a more stable and reliable avenue for wealth accumulation. On average, such strategic investments yield approximately 6-7 percent annual returns, underscoring the viability and attractiveness of a patient and disciplined investment approach.

```python
arr = np.array(data['Adj Close'])
index = data.index
prc_chng = ((arr[1:] / arr[:-1])-1)*100

plt.plot(pd.Series(prc_chng, index=index[1:]))
plt.title('Percentual Change')
plt.xlabel('Time')
plt.ylabel('% Change in asset')
plt.show()
```

# Comparative Analysis: Investment vs. Savings - Graphical Representation

In the following graph, we illustrate the evolution of our investment in contrast to the alternative of simply saving the money. The visual depiction reveals that in the short term, fluctuations may result in modest gains or occasional losses. However, the true value of the investment becomes apparent over the long term, where the market consistently outperforms traditional savings with an attractive annual interest rate.

It is noteworthy that the substantial increase in the investment becomes prominently visible in the final 3-4 years of the graph, emphasizing the efficacy of a patient and enduring investment strategy. This graphical representation underscores the enduring benefits of choosing the market over conventional savings for those seeking sustained financial growth.

```python
arr = np.array(inv_periods)
index = inv_periods.index
prc_chng = arr[1:] / arr[:-1]

prtfl = [every_month_inv]
for i in range(len(prc_chng)):
    app = prtfl[i]*prc_chng[i]+every_month_inv
    prtfl.append(app)
    
investment = np.array(list(range(1, len(prtfl)+1)))
investment = investment * every_month_inv
    
plt.plot(index, prtfl, color='green')
plt.plot(index, investment, color='red')
plt.title('Portfolio growth plotted, {}% return, {}% anualised'.format(perc_inv, perc_inv_ann))
plt.xlabel('Time')
plt.ylabel('Portfolio Worth in $')
plt.legend(['$ Invested', '$ Just Saved'], loc='best')
plt.show()
```
